Tasks

 1. Analyze content growth over year

```python
content_growth = df.groupby(['type', 'release_year'])['rating'].value_counts()
content_growth
```

2. Compare Movies vs TV Shows

```python
compare = df.groupby('type')['rating'].value_counts()
compare
```

3. Analyze country-wise content production


```python
country_wise = df.groupby(['country', 'type']).size().reset_index(name='count')
country_wise
```

4. Study content ratings

```python
content = df.groupby(['type', 'rating']).size().reset_index(name='count')
content
```

5. Identify dominant genres.


```python
dominant_genres = df.groupby('listed_in')['rating'].value_counts()
dominant_genres
```
